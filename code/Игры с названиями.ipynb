```python
# Определяем путь к файлу
file_path = r"C:\Users\rusla\OneDrive\Рабочий стол\Текстовый документ (3).txt"

# Функция для преобразования текста
def capitalize_after_last_dot(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    updated_lines = []
    
    for line in lines:
        # Удаляем символ переноса строки для обработки
        stripped_line = line.rstrip('\n')
        
        # Находим позицию последней точки
        last_dot_index = stripped_line.rfind('.')
        
        if last_dot_index != -1:
            # Разделяем строку на две части: до и после последней точки
            before_last_dot = stripped_line[:last_dot_index + 1]
            after_last_dot = stripped_line[last_dot_index + 1:]
            
            # Делаем первую букву после последней точки заглавной, если она существует
            if after_last_dot:
                updated_line = before_last_dot + after_last_dot.strip().capitalize()
            else:
                updated_line = stripped_line  # Если ничего нет после точки
            
            updated_lines.append(updated_line + '\n')  # Добавляем перенос строки
        else:
            # Если точки нет, оставляем строку без изменений
            updated_lines.append(line)  # Добавляем строку с переносом

    # Сохраняем обновленный текст в файл
    with open(file_path, 'w', encoding='utf-8') as file:
        file.writelines(updated_lines)

# Запускаем функцию
capitalize_after_last_dot(file_path)
```
